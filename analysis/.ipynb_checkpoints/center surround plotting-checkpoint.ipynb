```python
sweep_response = pd.read_hdf(analysis_file_path, 'sweep_response')
```

```python
expt_path = r'/Users/saskiad/Dropbox/Openscope Multiplex/Center Surround/Center_Surround_'+str(989418742)+'_data.h5'
stim_table = pd.read_hdf(expt_path, 'center_surround')
```

```python
stim_table['condition'] = 'ortho'
stim_table.loc[stim_table.Center_Ori==stim_table.Surround_Ori, 'condition'] = 'iso'
stim_table.loc[np.isfinite(stim_table.Center_Ori)&np.isnan(stim_table.Surround_Ori), 'condition'] = 'center'
stim_table.loc[np.isnan(stim_table.Center_Ori)&np.isnan(stim_table.Surround_Ori), 'condition'] = 'blank'
stim_table.loc[np.isnan(stim_table.Center_Ori)&np.isfinite(stim_table.Surround_Ori), 'condition'] = 'surround'
```

```python
plt.figure(figsize=(6,5))
plt.plot(sweep_response[(stim_table.condition=='center')&(stim_table.Center_Ori==90)]['199'].mean(), color='k')
plt.plot(sweep_response[(stim_table.condition=='ortho')&(stim_table.Center_Ori==90)]['199'].mean(), color='b')
plt.plot(sweep_response[(stim_table.condition=='iso')&(stim_table.Center_Ori==90)]['199'].mean(), color='r')
# plt.plot(sweep_response[(stim_table.condition=='surround')&(stim_table.Surround_Ori==90)]['199'].mean(), color='purple')
plt.axvspan(30,90, color='gray', alpha=0.1)
plt.tick_params(labelsize=16)
plt.xticks([30,60,90,120],[0,1,2,3])
plt.xlabel("Time (s)", fontsize=18)
plt.ylabel("DFF", fontsize=18)
plt.text(110,1.15, "center", color='k', fontsize=14)
plt.text(110,1.08, "iso", color='r', fontsize=14)
plt.text(110,1.01, "ortho", color='b', fontsize=14)
sns.despine()
```
